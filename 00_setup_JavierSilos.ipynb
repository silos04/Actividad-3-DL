```python
import hashlib

# El nombre debe coincidir EXACTAMENTE con el archivo que subiste
file_name = "00_setup_JavierSilos.ipynb"
sha256_hash = hashlib.sha256()

try:
    with open(file_name,"rb") as f:
        for byte_block in iter(lambda: f.read(4096),b""):
            sha256_hash.update(byte_block)

    hash_calculado = sha256_hash.hexdigest()
    print(f"Archivo: {file_name}")
    print(f"SHA-256: {hash_calculado}")

    # Guardamos el hash para usarlo en la siguiente celda
    %store hash_calculado

except FileNotFoundError:
    print(f"ERROR: Archivo '{file_name}' no encontrado.")
    print("Asegúrate de haberlo subido al panel de Archivos de Colab (Paso 3.2).")
```

```python
import json
import datetime

# Recuperamos el hash calculado en la celda anterior
%store -r hash_calculado

if 'hash_calculado' not in locals():
    hash_calculado = "ERROR_EJECUTA_LA_CELDA_ANTERIOR_PRIMERO"

registro = {
    "archivo": file_name, # Reutiliza el nombre de la celda anterior
    "sha265": hash_calculado,
    "fecha": datetime.datetime.now().isoformat()
}

file_json = 'registro_hash.json'
with open(file_json, 'w') as f:
    json.dump(registro, f, indent=4)

# Muestra el contenido del archivo creado
print(f"--- Contenido de {file_json} ---")
!cat {file_json}
```
